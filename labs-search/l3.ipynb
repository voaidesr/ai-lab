## Exercițiul 1

Rezolvați problema blocurilor cu bfs/ds, folosind clasele Nod și Graf discutate până acum

```python
from __future__ import annotations

class Nod:
    def __init__(self, stive, parinte : Nod | None = None, cost : int = 0):
        self.parinte = parinte
        self.stive = stive
        self.cost = cost

    def __eq__(self, other : Nod) -> bool:
        return self.stive == other.stive

    def __str__(self) -> str:
        return str(self.stive)

    def __repr__(self) -> str:
        return str(self.stive)

    def drumRadacina(self) -> list[Nod]:
        nod_curent = self
        drum = []
        while nod_curent is not None:
            drum.insert(0, nod_curent)
            nod_curent = nod_curent.parinte
        return drum

class Graf:
    def __init__(self, root: Nod, scop: Nod):
        self.root = root
        self.scop = scop

    def is_scop(self, nod: Nod) -> bool:
        return nod == self.scop

    def succesori(self, nod : Nod) -> list[Nod]:
        lista_succesori = []

        numar_stive = len(nod.stive)

        for idx in range(numar_stive):
            if not nod.stive[idx]:
                continue

            stiva = [list(elem) for elem in nod.stive]
            bloc = stiva[idx].pop()

            for j in range(numar_stive):
                if idx == j:
                    continue

                stiva_noua = [list(elem) for elem in stiva]
                stiva_noua[j].append(bloc)
                nod_nou = Nod(stiva_noua, nod, nod.cost + bloc)
                lista_succesori.append(nod_nou)

        return lista_succesori

n = Nod([[1], [2, 3], [4]])
g = Graf(n, n)
print(g.succesori(n))
```

```python
from collections import deque

def bfs(graf : Graf, n : int = 1) -> list[Nod]:
    radacina = graf.root
    q = deque([radacina])

    result = []

    while n > 0 and len(q) > 0:
        nod_curent = q.popleft()

        if graf.is_scop(nod_curent):
            result.append(nod_curent)
            n -= 1
            continue

        succesori = graf.succesori(nod_curent)
        q.extend(succesori)

    return result
```

```python
n = Nod([[1], [2, 3], [4]])
scop = Nod([[2, 3], [], [1, 4]])
g = Graf(n, scop)

drum = bfs(g)[0].drumRadacina()

for nod in drum:
    print(nod)
```

```python
def dfs(graf: Graf, n : int = 1) -> list[Nod]:
    radacina = graf.root
    result = []

    def dfs_recursiv(nod_curent: Nod):
        nonlocal n, result

        if n <= 0:
            return

        if graf.is_scop(nod_curent):
            result.append(nod_curent)
            n -= 1
            return

        succesori = graf.succesori(nod_curent)
        for succesor in succesori:
            if n > 0 and succesor not in nod_curent.drumRadacina():
                dfs_recursiv(succesor)

    dfs_recursiv(radacina)

    return result
```

```python
drum = dfs(g)[0].drumRadacina()

for nod in drum:
    print(nod)
```

Spunem că o euristică este **admisibilă** dacă valoarea estimării este mai mică sau egală decât costul celui mai scurt drum de la nodul respectiv la oricare dintre nodurile scop.

Spunem că o euristică $\hat{h}$ este **consistentă** dacă este monotonă:  $\hat{h}(n_i) \leq cost(n_i \rightarrow n_j) + \hat{h}(n_j)$, pentru orice $\{n_i, n_j\}$ aparținând grafului de stări, unde $n_j$ este succesor direct al lui $n_i$


Fiecare euristică va avea propria ei funcție, care va fi setată când rulăm prima dată programul.

\hat{h}(A) = 2, \quad \hat{h}(B) = 0
$$

$$
\hat{h}(A) = 2 \leq \mathrm{cost}(A, B) + \hat{h}(B) = 1 + 0 = 1 
$$

Ceea ce este arată că $h$ este nexonsistentă.
